```python
with open('mostPostingsRes.txt', 'r') as file: # reading from most postings MapReduce
    lines = file.readlines()

org_names = []
counts = []

for line in lines:
    parts = line.strip().split('\t')
    org_names.append(parts[0].strip('"'))  # Remove leading and trailing quotes
    counts.append(int(parts[1]))

# Select top N organizations
top_n = 30
top_org_names = org_names[:top_n]
top_counts = counts[:top_n]

plt.figure(figsize=(10, 6))
plt.barh(top_org_names, top_counts, color='skyblue')
plt.xlabel('Count')
plt.ylabel('Organization')
plt.title('Top {} Organizations by Postings'.format(top_n))
plt.gca().invert_yaxis()  # Invert y-axis to show the organization with the highest count at the top
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
