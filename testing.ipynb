```python
newdf = df[['name','position','html','orgCompany','orgAddress','orgTags','salary']]
print(newdf['position'] )
#find salary from the html field and print them 
print(newdf['html'][0])
print(newdf['html'][1])
print(newdf['html'][2])
# print(newdf['html'][0].find("salary"))
# print(newdf['html'][1].find("salary"))
# print(newdf['html'][2].find("salary"))
found_count=0
for i in range(0,numberOFrecords):
    if newdf['html'][i].find("salary") > 0:
        print(newdf['html'][i])
        print("found")
        found_count+=1
    else:
        print("not found")
print("found count",found_count)      




```
