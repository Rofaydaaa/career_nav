```python
import spacy

# Load the spaCy English tokenizer
nlp = spacy.load("en_core_web_sm")

job_description = """
The ideal candidate will develop and maintain software applications according to specifications. 
They will also coordinate with other teams to ensure project success and compliance.
They are expected to provide ongoing technical support and troubleshoot as necessary.
"""

# Process the job description with spaCy
doc = nlp(finaldf['html'][0])

# Extract sentences with action verbs often related to job duties
verbs_of_interest = {'develop', 'coordinate', 'provide', 'ensure', 'manage','require','maintain','support','troubleshoot'}
duties = []

for sent in doc.sents:
    # Check if any verb in the sentence is in our set of interest
    if any(token.lemma_ in verbs_of_interest for token in sent if token.pos_ == 'VERB'):
        duties.append(sent.text.strip())

print("Extracted Job Duties:")
for duty in duties:
    print(duty)
```

```python
import spacy
def get_required_skills(job_description):
    # Load the English NLP model
    nlp = spacy.load("en_core_web_sm")

    # Sample job description
    text = job_description   # Sample job description

    doc = nlp(text)

    i=0
    list_of_verbs_nouns = []
    while i < len(doc):
        if doc[i].pos_ == 'VERB':
            temp = []
            temp.append(doc[i])
            i+=1
            while i < len(doc) and doc[i].pos_ == 'NOUN':
                temp.append(doc[i])
                i+=1
            list_of_verbs_nouns.append(temp)
        else:
            i+=1

    # print(token.text, token.pos_)
    # Print the verbs and nouns
    # print(verbs_and_nouns)
    filtered_list =[]
    for i in list_of_verbs_nouns:
        if len(i) > 1:
            filtered_list.append(i)
    return filtered_list

# Get the required skills from the job description 
list_of_skills = []       
for i in range(numberOFrecords):
    print("record number: ",i,",,,,out of:",numberOFrecords-i)
    temp = []
    temp=get_required_skills(finaldf['html'][i])
    list_of_skills.append(temp)
```

```python
print("Extracted Required Skills:")
list_of_skills_to_string = []   
for i in list_of_skills:
    temp = []
    for j in i:
        temp.append(' '.join([token.text for token in j]))
    list_of_skills_to_string.append(temp)
list2=[]
for i in list_of_skills_to_string:
        list2.append(str(i))
skills_df = pd.DataFrame(list2,columns=['skills'])
finaldf= pd.concat([finaldf,skills_df],axis=1)
print(finaldf.head())
finaldf.to_csv('finaldf.csv',index=False)
```
