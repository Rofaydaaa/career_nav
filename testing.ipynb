```python
print(newdf['name'][0],"name")
print(newdf['position'][0],"position")
print(newdf['html'][0],"html")
print(newdf['orgCompany'][0],"orgCompany")
print(newdf['orgAddress'][0],"orgAddress")
print(newdf['orgTags'][0],"orgTags")
#carrer level done 
#salary done 
#work type/fulltime/partime/other-> consider it full time done 
#department done 
#country done 
#state done 
#company done 
#years of experince done 


#first let us find the carrer level 
found_count=0
#chec if carrer level exist in the position field which is a dicrionary
#create a new data frame with the carrer level if you find carer level exists then add it as is if not add NA




#here we will get the carrer level from the position field
careerLevel = []
for i in range(0,numberOFrecords):
    if "careerLevel" in newdf['position'][i]:
        if "months" in newdf['position'][i]['careerLevel']:
            careerLevel.append("JUNIOR")   
        elif "years" in newdf['position'][i]['careerLevel']:
            careerLevel.append("SENIOR")
        else:
            careerLevel.append(newdf['position'][i]['careerLevel'])
       
        #add to a new data frame
        found_count+=1
    else:
        careerLevel.append("NOT FOUND")

print("found count",found_count)
#turn the career level list into a data frame
careerLevel_df = pd.DataFrame(careerLevel,columns=['careerLevel'])
newdf= pd.concat([newdf,careerLevel_df],axis=1)


finaldf = newdf[['name','salary','careerLevel']]#will keep adding the rest of the columns
print(finaldf.head())   





#find work type key in the position field
found_count=0
workType = []
for i in range(0,numberOFrecords):
    if "workType" in newdf['position'][i]:
        workType.append(newdf['position'][i]['workType'])
        #add to a new data frame
        found_count+=1
    else:
        workType.append("NOT FOUND")
print("found count work type",found_count)
#turn the workType list into a data frame
workType_df = pd.DataFrame(workType,columns=['workType'])
finaldf= pd.concat([finaldf,workType_df],axis=1)
print(finaldf.head())





#find department key in the position field
found_count=0
department = []
for i in range(0,numberOFrecords):
    if "department" in newdf['position'][i]:
        department.append(newdf['position'][i]['department'])
        #add to a new data frame
        found_count+=1
    else:
        department.append("NOT FOUND")
print("found count department",found_count)
#turn the department list into a data frame
department_df = pd.DataFrame(department,columns=['department'])
finaldf= pd.concat([finaldf,department_df],axis=1)
print(finaldf.head())

#find country key in the orgAddress field
found_count=0
country = []
for i in range(0,numberOFrecords):
    if "country" in newdf['orgAddress'][i]:
        country.append(newdf['orgAddress'][i]['country'])
        #add to a new data frame
        found_count+=1
    else:
        country.append("NOT FOUND")
print("found count country",found_count)
#turn the country list into a data frame
country_df = pd.DataFrame(country,columns=['country'])
finaldf= pd.concat([finaldf,country_df],axis=1)
print(finaldf.head())

#find state key in the orgAddress field
found_count=0
state = []
for i in range(0,numberOFrecords):
    if "state" in newdf['orgAddress'][i]:
        state.append(newdf['orgAddress'][i]['state'])
        #add to a new data frame
        found_count+=1
    else:
        state.append("NOT FOUND")
print("found count state",found_count)

#turn the state list into a data frame
state_df = pd.DataFrame(state,columns=['state'])
finaldf= pd.concat([finaldf,state_df],axis=1)
print(finaldf.head())

#find name key in the orgCompany field
found_count=0
company = []
for i in range(0,numberOFrecords):
    if "name" in newdf['orgCompany'][i]:
        company.append(newdf['orgCompany'][i]['name'])
        #add to a new data frame
        found_count+=1
    else:
        company.append("NOT FOUND")
print("found count company",found_count)

#turn the company list into a data frame
company_df = pd.DataFrame(company,columns=['company'])
finaldf= pd.concat([finaldf,company_df],axis=1)
print(finaldf.head())


#find the word years count  in the html field
found_count=0
yearsOfExperince = []
for i in range(0,numberOFrecords):
    if "years" in newdf['html'][i]:
        #find the index of the word years in the html field
        years_index = newdf['html'][i].find("years")
        #get the years of experince value
        yearsValue=newdf['html'][i][years_index-2]
        if yearsValue.isdigit():
            yearsOfExperince.append(yearsValue)
            found_count+=1
        else:
            yearsOfExperince.append("NOT FOUND")
        #add to a new data frame
    else:
        yearsOfExperince.append("NOT FOUND")
print("found count yearsOfExperince",found_count)
#turn the yearsOfExperince list into a data frame
yearsOfExperince_df = pd.DataFrame(yearsOfExperince,columns=['yearsOfExperince'])
finaldf= pd.concat([finaldf,yearsOfExperince_df],axis=1)
print(finaldf.head())
finaldf.to_csv('finaldf.csv',index=False)


#concartenate the data frames html from newdf and to finaldf
finaldf = pd.concat([finaldf,newdf['html']],axis=1)
print(finaldf.head())











```
