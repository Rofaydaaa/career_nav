```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, FloatType

# Create a Spark session
spark = SparkSession.builder \
    .appName("Pandas to PySpark") \
    .getOrCreate()
```

```python
import pandas as pd

file_path = "/kaggle/input/job-preprocessed/final_output_processed_33k.csv"

# Read the Excel file into a Pandas DataFrame
df = pd.read_csv(file_path) 

print(df.head())
```

```python
# Define the schema for the PySpark DataFrame
schema = StructType([
    StructField("id", StringType(), nullable=False),
    StructField("sourceCC", StringType(), nullable=False),
    StructField("source", StringType(), nullable=False),
    StructField("locale", StringType(), nullable=False),
    StructField("position_name", StringType(), nullable=False),
    StructField("position_workType", StringType(), nullable=False),
    StructField("position_careerLevel", StringType(), nullable=False),
    StructField("position_department", StringType(), nullable=False),
    StructField("orgAddress_addressLine", StringType(), nullable=False),
    StructField("orgAddress_level", StringType(), nullable=False),
    StructField("orgAddress_country", StringType(), nullable=False),
    StructField("orgAddress_state", StringType(), nullable=False),
    StructField("orgAddress_city", StringType(), nullable=False),
    StructField("orgCompany_nameOrg", StringType(), nullable=False),
    StructField("orgCompany_homepage", StringType(), nullable=False),
    StructField("url", StringType(), nullable=False),
    StructField("orgTags_CATEGORIES", StringType(), nullable=True),
    StructField("orgTags_REQUIREMENTS", StringType(), nullable=True),
    StructField("orgTags_SKILLS", StringType(), nullable=True),
    StructField("salary_text", StringType(), nullable=False),
    StructField("salary", StringType(), nullable=True),
    StructField("years_or_months_experience", StringType(), nullable=False)
])

# Convert Pandas DataFrame to PySpark DataFrame using the specified schema
spark_df = spark.createDataFrame(df, schema=schema)

# Show the PySpark DataFrame
spark_df.show()
```
